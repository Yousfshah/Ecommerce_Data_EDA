### **Context**

This dataset represents the largest retail e-commerce orders dataset from Pakistan, covering a time span from March 2016 to August 2018. It consists of half a million transaction records, collected from various e-commerce merchants as part of a research study.

The dataset was released as a capstone project for the Data Science course at Alnafi. It provides valuable insights into the emerging e-commerce potential in Pakistan and serves as a resource for startups, researchers, and students to explore trends, challenges, and opportunities in the local e-commerce market.
Content

-    Geography: Pakistan
-    Time Period: March 2016 ‚Äì August 2018
-    Unit of Analysis: E-commerce orders
-    Size: 101 MB
-    File Type: CSV

Key Features:

- item_id: Unique identifier for each item in the dataset.
  
- status: Status of the order (e.g., complete, canceled, refunded).
  
- created_at: Date when the order was created.
  
- sku: Stock Keeping Unit, a unique code identifying products.
  
- price: Unit price of the item.
  
- qty_ordered: Quantity of the item ordered.
  
- grand_total: Total revenue generated by the item (price √ó quantity).
  
- increment_id: Unique order identifier for transactions.
  
- category_name_1: Primary product category associated with the item.
  
- sales_commission_code: Commission code related to the sale of the item.
  
- discount_amount: Discount applied to the item.
  
- payment_method: Payment method used for the transaction (e.g., COD, credit card).
  
- Working Date: The date when the order was processed.
  
- BI Status: Business Intelligence status or categorization of the transaction.
  
- MV: Market value of the item.
  
- Year: Year of the transaction.
  
- Month: Month of the transaction.
  
- Customer Since: The starting date of the customer relationship.
  
- M-Y: Month and year representation (e.g., 7-2016).
  
- FY: Fiscal year (e.g., FY17).
  
- Customer ID: Unique identifier for the customer.
  
- Unnamed: 21 to Unnamed: 25: Empty columns with no data; likely irrelevant and should be dropped.

### **Purpose**

This dataset enables analysis of Pakistan‚Äôs e-commerce sector, uncovering:

-    Trends: Seasonal and product-wise.
-    Customer Behavior: Payment preferences and shopping habits.
-    Performance: Top-performing product categories and payment methods.
-    Business Insights: Identifying growth opportunities for startups and established e-commerce platforms.

```python
# remove Unnecessary columns
df.drop(['Unnamed: 21', 'Unnamed: 22', 'Unnamed: 23', 'Unnamed: 24', 'Unnamed: 25'], axis=1, inplace=True)
print("Unnecessary columns ('Unnamed: 21, to 'Unnamed: 25') have been successfully dropped from the dataset")
```

```python
# I will drop the 'sales_commission_code' column due to over 50% missing values, as it is not suitable for filling due to its uniqueness
df.drop('sales_commission_code', axis=1, inplace=True)
print("'sales_commission_code' column has been successfully dropped from the dataset")
```

```python
# Check Missing Values
percentage_missing = (df.isnull().sum() / len(df)) * 100
percentage_missing = percentage_missing.map("{:.0f}%".format)
percentage_missing
```

### Let's deal with missing values

```python
# Our data has so many rows with entirely NaN values so we have to remove 
df.tail(8)
```

```python
# it will remove rows where all records are empty
df.dropna(how='all', inplace=True)
print("Rows with all values missing have been successfully dropped from the dataset")
```

```python
# now see missing values
df.isnull().sum() 
```

```python
# Fill missing values in 'status', 'category_name_1', and 'Customer Since' with their respective mode 
# because These columns contain categorical data
df['status'].fillna(df['status'].mode()[0], inplace=True)
df['category_name_1'].fillna(df['category_name_1'].mode()[0], inplace=True)
df['Customer Since'].fillna(df['Customer Since'].mode()[0], inplace=True)

# Drop rows where 'sku' or 'Customer ID' have missing values
# becuase 'sku' and 'Customer ID' are unique 
df.dropna(subset=['sku', 'Customer ID'], inplace=True)

# Display a message confirming the operation
print("Missing values filled for 'status', 'category_name_1', and 'Customer Since'.")
print("Rows with missing values in 'sku' or 'Customer ID' have been dropped.")
```

```python
# successfully we deal with missing values
df.isnull().sum()
```

```python
# handle data inconsistency

# Remove 'FY' from the column name FY
df['FY'] = df['FY'].str.replace('FY', '', regex=True)

# Rename columns with meaningful names
df.rename(columns={
    ' MV ': 'market_value',
    'M-Y': 'month_year',
    'FY': 'fiscal_year',
    'sku': 'stock_keeping_unit', 
    'Customer Since': 'customer_since',
    'category_name_1': 'category',
    'BI Status': 'bi_status',
    'Working Date': 'working_date'
}, inplace=True)


# Our dataset has \\N So Replace '\\N' with NaN and drop rows containing NaN values
df.replace('\\N', pd.NA, inplace=True)  # Convert '\\N' to NaN
df.dropna(how='any', inplace=True)  # Drop rows that contain NaN in any column

print("Handle Data Inconsistency completed successfully")
```

```python
# Fix Data Types
# as we mention above in data exploration part 


# 1. Convert date-related columns to datetime
date_columns = ['working_date', 'created_at', 'customer_since', 'Year', 'Month', 'month_year', 'fiscal_year']
df[date_columns] = df[date_columns].apply(pd.to_datetime, errors='coerce')

# 2. Convert specified columns into categorical type
categorical_columns = ['status', 'payment_method', 'category', 'bi_status', 'stock_keeping_unit']
df[categorical_columns] = df[categorical_columns].astype('category')

# 3. Convert specified columns into integer type
integer_columns = ['price', 'qty_ordered', 'Customer ID', 'increment_id', 'grand_total', 'item_id', 'market_value', 'discount_amount']
df[integer_columns] = df[integer_columns].apply(pd.to_numeric, errors='coerce').astype(int)

print("Data type conversion process completed successfully")
```
